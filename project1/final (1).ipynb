```python
%load_ext rmagic
%load_ext rpy2.ipython
```

```python
%R wordlist <- ""
```

open file

```python
%R f <- file("https://s3.amazonaws.com/2017-dmfa/project-1/holmes.txt")
```

```python
%R open(f)
```

split lines of text into one word per line

```python
%R while(length(line <- readLines(f, n=1)) > 0)  wordlist <- c(wordlist,unlist(strsplit(line,"\\W+|\\d+")))
```

Transform  word into lower case

```python
%R lowerword <- tolower(grep("[a-z]|[A-Z]",wordlist,value = TRUE))
```

```python
%R print(sort(table(lowerword),decreasing=T)[1:10])
```
